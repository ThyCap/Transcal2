#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Initializing variables

```python
# Geometry
L = 1
Nx = 100
dx = L/(Nx - 1)

# Time
dt = 0.001
Nt = 5000

# Material
alpha = 0.02

# Heat Source
Q = 0

# Boundary Conditions
T_left = 0
T_right = 100

coef = alpha*dt/(dx**2)
```

#### Setup and calculation

```python
T = np.zeros((Nt, Nx))
```

```python
# Initializing BCs
T[0] = np.full(Nx, T_left, dtype= float)
T[0][-1] = T_right
```

```python
coef
```

```python
for i in range(Nt):
    for j in range(Nx):
        if j == 0:
            T[i,j] = T_left
        elif j == Nx - 1:
            T[i,j] = T_right
        else:
            T[i,j] = T[i - 1][j] + (T[i - 1][j-1] - 2*T[i - 1][j] + T[i - 1][j+1])*coef
```

```python
plt.plot(np.linspace(0, L, Nx), T[0])
plt.plot(np.linspace(0, L, Nx), T[50])
plt.plot(np.linspace(0, L, Nx), T[100])
plt.plot(np.linspace(0, L, Nx), T[150])
plt.plot(np.linspace(0, L, Nx), T[200])
plt.plot(np.linspace(0, L, Nx), T[250])
plt.plot(np.linspace(0, L, Nx), T[300])
plt.plot(np.linspace(0, L, Nx), T[350])
plt.plot(np.linspace(0, L, Nx), T[400])
plt.plot(np.linspace(0, L, Nx), T[450])

plt.show()
```
